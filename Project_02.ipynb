Высоту потолков средним значением будет заменять не правильно, поскольку это важный параметр для многих при выборе квартиры.
Высота потолков в Санкт-Петербурге часто зависит от удаления от центра (так в центре это часто 3 м, а на окраине ближе к 2.5 (эконом-класс)).
В соответствии с этой теорией напишем функцию, которая присвоит наименование каждой категории(в зависимости от дальности центра, где это возможно) и уже по категориям заменим средним значением.

Введем следующие категории:
- 0-2км от центра (категория 'центр');
- 2-5км от центра (категория 'близко к центру');
- 5-17км от центра (категория 'доступность центра');
- больше 17км от центра (категория 'удаленное жилье').

Так выглядит логично: 0-2,5 км - квартира у Дворцовой, 2-5 км - квартира на ВО, 5-17км - квартира на Парнасе или в Купчино. > 17км - за КАДом.


Напишем функцию, определяющую удаленность от центра

```python
def fillna_ceiling_height(row):
    if row['city_centers_nearest'] < 2500:
        return 'центр'
    elif row['city_centers_nearest'] < 5000:
        return 'близко к центру'
    elif row['city_centers_nearest'] < 17000:
        return 'доступность центра'
    elif row['city_centers_nearest'] >= 17000:
        return 'удаленное жилье'
```

```python
df['from_center_category'] = df.apply(fillna_ceiling_height, axis=1)
```

```python
df[['locality_name', 'city_centers_nearest', 'from_center_category']].head(10)
```

Заполним пропуски в столбце ceiling_height средним по каждой группе from_center_category (кроме строк, где NaN в city_centers_nearest - там мы не можем узнать, центр это или, например, почти область, поэтому не будем ничего делать)

Посмотрим среднее по каждой категории, заодно проверим нашу теорию, что при удалении от центра высота потолков уменьшается

```python
df.groupby('from_center_category')['ceiling_height'].mean().sort_values(ascending=False).to_frame()
```

теория о том, что высота потолков уменьшается при удаленности от, центра подтвердилась.
